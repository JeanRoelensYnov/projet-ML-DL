```python
print(data.size)
print(data.drop_duplicates().size)

testCount = data.groupby('track_id')['track_genre'].nunique().to_frame().reset_index()
print(testCount[testCount['track_genre']>1])
#Un track_id correspond à plusieurs track_genre / UPDATE / CountVectorizer refuse de fonctionner avec le track_genre nous le retirons donc du dataset
```

```python
data_clean = data.drop(labels=['track_id', 'track_genre'], axis='columns')
data_clean = data_clean.drop_duplicates()
encoder=RobustScaler()
encoder_artist = LabelEncoder()
encoder_album = LabelEncoder()
encoder_track_name = LabelEncoder()

data_clean['artists'] = encoder_artist.fit_transform(data_clean['artists'])
data_clean['album_name'] = encoder_album.fit_transform(data_clean['album_name'])
data_clean['track_name'] = encoder_track_name.fit_transform(data_clean['track_name'])

data_encode = encoder.fit_transform(data_clean)
```

```python
data_clean.size
```

```python
data = p.DataFrame(data=data_clean, columns=data.columns)
```

```python
matr = data.corr()

plt.figure(figsize=(20, 16)) 
sns.heatmap((matr < -0.7) | (matr > 0.7), annot=True, fmt=".2f", cmap='coolwarm')  
plt.title('Matrice de Corrélation')  
plt.show()
```

```python
#seeing the matrix we cut the "energy" column has it is too correlate to "loudness" and "acousticness"
#we will drop the album too because we do not want to give this information to the model when guessing
data = data.drop(labels='track_id', axis='energy')
data = data.drop(labels='album_names', axis='energy')
```
